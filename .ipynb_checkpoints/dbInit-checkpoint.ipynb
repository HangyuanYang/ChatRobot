```python
import requests
import json
import sqlite3


def buildDB():
    url = "https://free-nba.p.rapidapi.com/games"

    querystring = {"page": "１", "per_page": "100"}

    headers = {
        'x-rapidapi-host': "free-nba.p.rapidapi.com",
        'x-rapidapi-key': "96d8bc25cemshc7c0f92a27c3617p1ac8c4jsn9f4487243965"
    }

    response = requests.request("GET", url, headers=headers, params=querystring)

    # print(response.text)
    # type(json.loads(response.text))
    result = json.loads(response.text)
    number = result['meta']["total_pages"]
    conn = sqlite3.connect('DB.db')
    c = conn.cursor()
    c.execute("DROP TABLE if exists game")
    c.execute(
        "CREATE TABLE IF NOT EXISTS "
        "game(id int, date text　COLLATE NOCASE, home_team int, home_team_score int, period int,postseason NUMERIC,"
        "season int, status text　COLLATE NOCASE,time text　COLLATE NOCASE, visitor_team int,visitor_team_score int)")

    for i in range(number):
        querystring = {"page": str(i+1), "per_page": "100"}

        headers = {
            'x-rapidapi-host': "free-nba.p.rapidapi.com",
            'x-rapidapi-key': "96d8bc25cemshc7c0f92a27c3617p1ac8c4jsn9f4487243965"
        }

        response = requests.request("GET", url, headers=headers, params=querystring)

        # print(response.text)
        # type(json.loads(response.text))
        result = json.loads(response.text)
        for game in result['data']:
            game["home_team"]=game["home_team"]["id"]
            game["visitor_team"]=game["visitor_team"]["id"]
            insert_str = "INSERT INTO game(id,date,home_team,home_team_score,period,postseason," \
                         "season,status,time,visitor_team,visitor_team_score) " \
                         "VALUES(?,?,?,?,?,?,?,?,?,?,?)"
            t = tuple(game.values())
            c.execute(insert_str, t)
        c.execute("commit")
        print("Gamepage ",i+1)

    return True


buildDB()
```

```python
import requests
import json
import sqlite3


def buildDB():
    url = "https://free-nba.p.rapidapi.com/teams"

    querystring = {"page": "1"}

    headers = {
        'x-rapidapi-host': "free-nba.p.rapidapi.com",
        'x-rapidapi-key': "96d8bc25cemshc7c0f92a27c3617p1ac8c4jsn9f4487243965"
    }

    response = requests.request("GET", url, headers=headers, params=querystring)

    # print(response.text)
    # type(json.loads(response.text))
    result = json.loads(response.text)

    conn = sqlite3.connect('DB.db')
    c = conn.cursor()
    c.execute("DROP TABLE if exists team")
    c.execute(
        "CREATE TABLE IF NOT EXISTS team(id int, abbreviation text　COLLATE NOCASE, city text　COLLATE NOCASE, "
        "conference text　COLLATE NOCASE, division text　COLLATE NOCASE,full_name text　COLLATE NOCASE,name text　COLLATE NOCASE)")
    for team in result['data']:
        insert_str = "INSERT INTO team(id, abbreviation, city, conference, division, full_name, name) VALUES(?,?,?,?,?,?,?)"
        t = tuple(team.values())
        c.execute(insert_str, t)
    c.execute("commit")
    return True


buildDB()
```
